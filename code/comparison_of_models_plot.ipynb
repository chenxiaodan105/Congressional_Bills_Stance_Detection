```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
have_stance = [0.21,0.16,0.22,0.19,0.80,0.88,0.86,0.86]
no_stance = [0.85,0.96,0.83,0.78,0.83,0.91,0.96,0.95]
index = ['TfIdf + Logistic Regression','TfIdf + Random Forest',
         'TfIdf + Balanced Random Forest','TfIdf + Weighted Random Forest',
         'TfIdf + Relabeling Model + Logistic Regression',
         'TfIdf + Relabeling Model + Random Forest',
         'Relabeling Model + GloVe + LSTM',
         'Relabeling Model + GloVe + CNN + LSTM']
df = pd.DataFrame({'have_stance': have_stance,
                   'no_stance': no_stance}, index=index)
df.plot.barh(width=.7,figsize=(16,10),fontsize=14,
                       title='Comparison of Combinations')
plt.xlabel('F1 score' )
plt.show()
```
